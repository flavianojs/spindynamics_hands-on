**1. Introduction: Noncollinear Magnetism**

Noncollinear magnetism refers to magnetic states in which spin moments are neither parallel nor antiparallel to each other but instead form finite angles, resulting in complex spatial configurations. These states emerge from the competition among various magnetic interactions, such as the Heisenberg exchange, the Dzyaloshinskii‚ÄìMoriya interaction (DMI), magnetic anisotropy, and coupling to external fields.

In ferromagnetic systems, all spins tend to align in the same direction, minimizing the exchange energy. In antiferromagnets, neighboring spins orient antiparallel, canceling the net magnetic moment. In contrast, *altermagnetism* represents a newly identified class of magnetic order characterized by alternating spin polarization on different crystal sublattices, combining symmetries typical of antiferromagnetism with transport effects similar to those in ferromagnetic materials.

**Ferromagnets vs. Antiferromagnets vs. Altermagnets**

![](https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcTuLm9ahCf0e9OYAd9ZLmVjKDGDWqBrK6Mxgg&s)

When multiple interactions compete ‚Äî for instance, in triangular or hexagonal lattices ‚Äî *magnetic frustration* arises, preventing the simultaneous satisfaction of all exchange interactions. This frustration naturally leads to noncollinear and noncoplanar states, where spins form specific angles with one another to minimize the total energy.

The Dzyaloshinskii‚ÄìMoriya interaction plays a crucial role in stabilizing noncollinear and chiral textures, such as *spin spirals* and *magnetic skyrmions*. Spin spirals consist of a continuous rotation of spins along a spatial direction, whereas skyrmions are stable topological structures in which spins form a vortex-like configuration with nontrivial topology.

# 2. Heisenberg Model

The **Heisenberg model** is one of the fundamental models in the physics of magnetism. It describes how the magnetic moments (or spins) of atoms interact with each other within a material. In this model, each atom is represented by a spin vector, and the total energy of the system depends on the relative alignment of these vectors.

The simplest form of the Heisenberg Hamiltonian is:

\begin{equation}
 H = -\sum_{i,j} J_{ij} \mathbf{S}_i \cdot \mathbf{S}_j
\end{equation}

where $ \mathbf{S}_i $ and $ \mathbf{S}_j $ are the spin vectors at sites $ i $ and $ j $, and $ J_{ij} $ is the exchange parameter, which indicates the strength and nature of the interaction between spins.
If $ J_{ij} > 0 $, the system tends to align the spins in the same direction (**ferromagnetic interaction**);
if $ J_{ij} < 0 $, the spins prefer to align in opposite directions (**antiferromagnetic interaction**).

This model is essential for understanding the magnetic properties of solids, such as ferromagnetism and antiferromagnetism, and serves as the starting point for more complex theories that include quantum effects, anisotropies, and additional interactions.

---

## Dzyaloshinskii‚ÄìMoriya Interaction, Magnetic Anisotropy, and Zeeman Energy

These interactions are usually added as **additional terms to the Heisenberg Hamiltonian** in order to describe more realistically the behavior of real magnetic materials. They introduce new coupling mechanisms between spins, responsible for a wide variety of experimentally observed magnetic properties and textures.

The **Dzyaloshinskii‚ÄìMoriya interaction (DMI)**, **magnetic anisotropy**, and **Zeeman energy** are therefore fundamental elements for understanding the complex behavior of magnetic systems, especially those with **noncollinear** or **chiral spin textures**.

---

## Dzyaloshinskii‚ÄìMoriya Interaction (DMI)

The **Dzyaloshinskii‚ÄìMoriya interaction** arises from spin‚Äìorbit coupling in systems that lack inversion symmetry. It favors a *tilted* alignment between neighboring spins, instead of the parallel or antiparallel configurations preferred by the Heisenberg exchange.

The DMI Hamiltonian is given by:

\begin{equation}
H_\text{DMI} = \sum_{i,j} \mathbf{D}_{ij} \cdot (\mathbf{S}_i \times \mathbf{S}_j),
\end{equation}

where **$\mathbf{D}_{ij}$** is the Dzyaloshinskii‚ÄìMoriya vector, which determines the sense of rotation between the spins.
This interaction is responsible for stabilizing **noncollinear** and **topologically nontrivial spin structures**, such as *skyrmions* and *spin spirals*.

---

## Magnetic Anisotropy

**Magnetic anisotropy** describes how the magnetic energy of a material depends on the direction of magnetization relative to its crystal lattice. It mainly originates from spin‚Äìorbit coupling and crystal field effects, leading to preferred directions ‚Äî called *easy axes* ‚Äî of magnetization.

The simplest form of the anisotropy energy is:

\begin{equation}
H_\text{ani} = -K (\mathbf{S} \cdot \hat{n})^2,
\end{equation}

where **$K$** is the anisotropy constant and **$\hat{n}$** defines the easy axis.

* If $(K > 0)$, spins tend to align along $\hat{n}$ (*easy-axis anisotropy*).
* If $(K < 0)$, spins prefer to align in the plane perpendicular to $\hat{n}$ (*easy-plane anisotropy*).

---

## Zeeman Energy

The **Zeeman energy** represents the interaction between magnetic moments and an external magnetic field $\mathbf{B}$. It is expressed as:

\begin{equation}
H_\text{Z} = - g \mu_B \sum_i \mathbf{B} \cdot \mathbf{S}_i,
\end{equation}

where **$g$** is the Land√© factor and **$\mu_B$** is the Bohr magneton.
This term tends to align the spins along the direction of the external magnetic field.

---

## Total Hamiltonian

Combining all these terms ‚Äî exchange interaction, DMI, anisotropy, and Zeeman energy ‚Äî the total Hamiltonian can be written as:

\begin{equation}
H = - \sum_{i,j} J_{ij}  \mathbf{S}_i \cdot \mathbf{S}_j
- \sum_{i,j} \mathbf{D}_{ij} \cdot (\mathbf{S}_i \times \mathbf{S}_j)
- K \sum_i (\mathbf{S}_i \cdot \hat{n})^2
- g \mu_B \sum_i \mathbf{B} \cdot \mathbf{S}_i
\end{equation}

In this Hamiltonian:

* The **first term** represents the **Heisenberg exchange**, which tends to align or antialign neighboring spins depending on the sign of $J_{ij}$;
* The **second term** is the **Dzyaloshinskii‚ÄìMoriya interaction**, which promotes a tilted orientation between spins and gives rise to chiral textures;
* The **third term** corresponds to **magnetic anisotropy**, which defines preferred directions of magnetization;
* The **last term** is the **Zeeman energy**, which aligns spins with the external magnetic field.

---

## Summary

These three interactions ‚Äî **DMI**, **magnetic anisotropy**, and **Zeeman coupling** ‚Äî compete and cooperate with the **Heisenberg exchange**, resulting in a wide variety of fundamental magnetic states and excitations.
Their interplay underlies many modern phenomena in **spintronics**, **chiral magnetism**, and **topological magnetic materials**.





# 3. Equa√ß√£o de Landau‚ÄìLifshitz‚ÄìGilbert (LLG): din√¢mica de spin

A **equa√ß√£o de Landau‚ÄìLifshitz‚ÄìGilbert (LLG)** descreve a **din√¢mica temporal dos momentos magn√©ticos** em um material. Ela captura como o vetor de magnetiza√ß√£o $ \mathbf{M} $ (ou spin $ \mathbf{S} $) evolui sob a a√ß√£o de campos efetivos internos e externos, incluindo tanto a **precess√£o** quanto os **efeitos dissipativos** que conduzem o sistema ao equil√≠brio.

A equa√ß√£o √© escrita como:

\begin{equation}
\frac{d\mathbf{M}_i}{dt} = -\gamma  \mathbf{M}_i \times \mathbf{H}_{i}^\text{eff} + \frac{\alpha}{M_s}  \mathbf{M}_i \times \frac{d\mathbf{M}_i}{dt},
\end{equation}

onde:

* $ \gamma $ √© a **raz√£o giromagn√©tica**, que relaciona o momento magn√©tico ao momento angular;
* $ \alpha $ √© o **par√¢metro de amortecimento de Gilbert**, respons√°vel pela dissipa√ß√£o de energia;
* $ M_s $ √© a **magnetiza√ß√£o de satura√ß√£o**;
* e $ \mathbf{H}^{\text{eff}}_i $ √© o **campo efetivo local** no s√≠tio $ i $.

---

### Campo efetivo no modelo de Heisenberg

No contexto do **modelo de Heisenberg**, o campo efetivo $ \mathbf{H}_{\text{eff},i} $ √© determinado pela derivada funcional do **Hamiltoniano magn√©tico total** em rela√ß√£o ao momento de spin local $ \mathbf{S}_i $:

\begin{equation}
\mathbf{H}_{i}^\text{eff} = -\frac{1}{\hbar \gamma} \frac{\partial H}{\partial \mathbf{S}_i}.
\end{equation}

Considerando o Hamiltoniano de Heisenberg total, temos:

\begin{equation}
\mathbf{H}_{i}^\text{eff} = \frac{1}{\hbar \gamma}
\left(
\sum_j J_{ij} \mathbf{S}_j
- \sum_j \mathbf{D}_{ij} \times \mathbf{S}_j
- 2 K_i (\mathbf{S}_i \cdot \hat{n}_i)\hat{n}_i
- \mu_B \mathbf{B}
  \right).
\end{equation}

Esse campo efetivo incorpora todas as contribui√ß√µes que influenciam a din√¢mica do spin.

Assim, a **equa√ß√£o de LLG**, combinada com a defini√ß√£o de $ \mathbf{H}^{\text{eff}} $ no modelo de Heisenberg, fornece uma descri√ß√£o completa e autoconsistente da **din√¢mica de spin** em materiais magn√©ticos ‚Äî desde o regime coletivo das ondas de spin (*magnons*) at√© as excita√ß√µes topol√≥gicas como *skyrmions* e *spin spirals*.

````python
import numpy as np

def read_spirit_input(filename):
  with open(filename, 'r') as f:
    lines = f.readlines()

    to_read_bravais = False
    counter_bravais = 0

    for line in lines:

      # read bravais vectors
      if 'bravais_vectors' in line:
        to_read_bravais = True
        continue
      if to_read_bravais:
        x = float(line.split()[0])
        y = float(line.split()[1])
        z = float(line.split()[2])
        counter_bravais += 1
        if counter_bravais == 1: a1 = np.array([x,y,z])
        if counter_bravais == 2: a2 = np.array([x,y,z])
        if counter_bravais == 3: a3 = np.array([x,y,z]); to_read_bravais = False

      #read n_basis_cells
      if 'n_basis_cells' in line:
        n1 = int(line.split()[1])
        n2 = int(line.split()[2])
        n3 = int(line.split()[3])
        n_basis_cells = np.array([n1, n2, n3])

  return a1, a2, a3, n_basis_cells

def create_position_lattice(a1, a2, a3, n_basis_cells, toprint=True):
  spin_positions = []

  for i in range(n_basis_cells[0])-n_basis_cells[0]/2:
    for j in range(n_basis_cells[1])-n_basis_cells[1]/2:
        for k in range(n_basis_cells[2])-n_basis_cells[2]/2:

          posicao = a1*i + a2*j + a3*k
          spin_positions.append(posicao)

  spin_positions = np.array(spin_positions)
  if toprint:
    print("Spin positions:")
    print(spin_positions)

  return spin_positions

def plotting_spin_lattice_aux(spin_positions, spin_vectors, figname='spirit_output.png', cameralookat=None, rescale=1):
  import gr, gr3
  import colorsys

  gr3.clear()

  colors = np.zeros(spin_positions.shape, dtype=np.float32)

  for i, vector in enumerate(spin_vectors):
      x = np.sqrt(np.square(vector[0])+np.square(vector[1]))
      hue =   np.arctan2(x,vector[2]) / np.pi / 1.5
      saturation = 1
      value = 1

      colors[i] = colorsys.hsv_to_rgb(hue, saturation, value)

  for i, position in enumerate(spin_positions):
      size_arrow = np.linalg.norm( spin_vectors[i])
      gr3.drawspins(4*spin_positions[i], spin_vectors[i], colors[i],
                    cone_radius=0.28*rescale*size_arrow, cylinder_radius=0.15*rescale*size_arrow,
                    cone_height=0.6*rescale*size_arrow, cylinder_height=0.7*rescale*size_arrow)

  gr3.setbackgroundcolor(1. , 1. , 1. , 0.5)
  # gr3.cameralookat(0, 0, 30, 0, 0, 0, 0, 1, 0) #Angle 1: Top view
  gr3.cameralookat(0, 1, 100, 0, 0, 0, 0, 0,-1) #Angle 2:
  if cameralookat is not None:
    gr3.cameralookat(*cameralookat)
  # gr3.setlightdirection(0.2,0.2,1)
  gr3.setlightdirection(*cameralookat[0:3])

  return gr3.export(figname, 500, 500)

def plotting_spin_lattice(spirit_input, spin_vectors=None, spirit_output=None, cameralookat=None, debug=False):
  # Reading Spirit input file
  a1, a2, a3, n_basis_cells = read_spirit_input(spirit_input)

  # Creating spin lattice
  spin_positions = create_position_lattice(a1, a2, a3, n_basis_cells, toprint=debug)

  # Reading spin vectors
  if spin_vectors is None:
    if spirit_output:
      spin_vectors = np.loadtxt(spirit_output, usecols=(0, 1, 2))
    else:
      raise ValueError("Please provide spin_vectors or spirit_output")

  if debug: print("Vectors:"); print(spin_vectors)

  figname = spirit_input[0:-4]+'.png'

  # Plotting 3D render of spin lattice
  return plotting_spin_lattice_aux(spin_positions, spin_vectors, figname, cameralookat, rescale=3)

def editing_file(filename, key, value, append='', replace = False, append_after = False, num_line = 0, block=False):
   # opening and reading the file
   file = open(filename, 'r')
   lines=file.readlines()
   file.close()

   # reopening the file in the writing mode
   target = open(filename, 'w')

   if append_after :
      num_line = max([1, num_line])
      replace = False

   # looping through the files looking for keywords
   done = False
   line_number = -1
   for i, line in enumerate(lines) :
      if key in line and not done :
         line_number = i
         done = True

      if done and i == line_number+num_line and not block :
         if num_line == 0 and not replace:
            newline = key + str(value) + append + '\n'
         else :
            newline = value + append + '\n'

         if not value == 'xxx':
            target.write( newline )
            minor = min(len(newline)-1, 100)
            print( 'Modification in "' + filename + '": ' + newline[0:minor])

            if not replace and not num_line == 0  :
               target.write( line )
      elif done and block:
         if i == line_number:
            target.write( line )
         elif i == line_number+1:
            target.write( value )
            minor = min(len(value)-1, 100)
            print( 'Modification in "' + filename + '": ' + value[0:minor] )
         elif i > line_number+1 and i < line_number + num_line+1:
            pass
         else:
            target.write( line )
            done = False
      else :
         target.write( line )

   if line_number+num_line > len(lines)-1:
      newline = value + append + '\n'
      target.write( newline )
      minor = min(len(newline)-1, 100)
      print( 'Modification in "' + filename + '": ' + newline[0:minor] )

   target.close()
````

````python
import matplotlib.pyplot as plt

def column_to_plot(folder, toprint=False):
    import math

    vampire_input = folder + '/input'

    if toprint: print('Input file:', vampire_input)

    # read vampire input and return all lines with "output:"
    outputs = []
    mat_file = None
    with open(vampire_input, 'r') as f:
        for line in f:
            if "output:" in line and line[0] != "#":
                outputs.append(line.split(':')[1])

            if "material:file" in line and line[0] != "#":
                mat_file = line.split('=')[1].replace('\n','')
                mat_file = vampire_input.split('/input')[0] + '/' + mat_file
                if toprint: print('Material file:', mat_file)

    if mat_file:
        with open(mat_file, 'r') as f:
            for line in f:
                if "material:num-materials" in line and line[0] != "#":
                    num_spins = int(line.split('=')[1])
                    if toprint: print('Number of spins:', num_spins, ', read from material file.')

    num_column_outputs = {
        "real-time":                   {"quantity size":1, "n_spin":1},
        "temperature":                 {"quantity size":1, "n_spin":1},
        "applied-field-strength":      {"quantity size":1, "n_spin":1},
        "material-mean-magnetisation": {"quantity size":4, "n_spin":num_spins},
        "material-mean-susceptibility":{"quantity size":4, "n_spin":num_spins},
        "material-mean-specific-heat": {"quantity size":1, "n_spin":num_spins},
        "mean-magnetisation":          {"quantity size":4, "n_spin":1},
        "mean-susceptibility":         {"quantity size":4, "n_spin":1},
        "mean-specific-heat":          {"quantity size":1, "n_spin":1},
    }

    num_output = 0
    list_of_output = {}
    # outputs_to_plot = []
    start_column = 0
    final_column = 0

    for i, output in enumerate(outputs):
        output_aux = output.replace('\n','')
        output_aux = output_aux.replace(' ','')

        quantity_size = num_column_outputs[output_aux]["quantity size"]
        n_spin =  num_column_outputs[output_aux]["n_spin"]
        start_column = final_column
        final_column = start_column + quantity_size*n_spin

        if "real-time" not in output and "temperature" not in output and "applied-field-strength" not in output:
            num_output+=1
            # outputs_to_plot.append({'output':output_aux, 'start_column':start_column, 'final_column':final_column, 'quantity size':quantity_size, 'n_spin':n_spin })
            list_of_output[output_aux] = { 'start_column':start_column, 'final_column':final_column, 'quantity size':quantity_size, 'n_spin':n_spin }

            if toprint: print('Output:', output_aux, 'columns:', start_column, '-', final_column)

        if "temperature" in output:
            first_column = start_column
            if toprint: print("Reading temperature in column:", first_column)
            label_xaxis = 'Temperature (K)'
        if "applied-field-strength" in output:
            first_column = start_column
            if toprint: print("Reading applied-field-strength in column:", first_column)
            label_xaxis = 'Applied field (T)'

    return list_of_output

# list_of_output = column_to_plot(folder='../vampire_tripleUC_3mat_DMI2/output-Field8-0T-Bc_poor_res_stat5', toprint=True)

def phase_diagram_colormap(folder, color='gray', property=36, ax=None, fig=None, title=None, **args):
    import os
    files = [f for f in os.listdir(folder) if f.startswith("output_") ]
    files.sort()

    if ax is None:
        fig, ax = plt.subplots()

    if property == 'susceptibility':
        cols = [36]
    elif property == 'specific_heat':
        cols = [37]
    else : cols = property

    if type(cols) == int:
        cols = [cols]

    if 'plot_derivative' in args:
        plot_derivative = args['plot_derivative']
    else :
        plot_derivative = False

    if 'xlim' in args:
        xlim = args['xlim']
    else :
        xlim = False

    if 'cmap' in args:
        cmap = args['cmap']
    else :
        cmap = 'viridis'

    if 'shading' in args:
        shading = args['shading']
    else :
        shading = 'gouraud'

    Ts = []
    zs = np.array([])
    for file in files:
        # if file == files[0]: continue # if to skip the first temperature
        temperature = float(file.split('_')[1])
        Ts.append(temperature)
        z_sum=0
        for column in cols:
            Bs, z = np.loadtxt(folder + '/' + file, usecols=(1,column)).T
            z_sum += z

        # find index of element 8 in Bs
        idx = np.where(Bs == max(Bs))[0][0]
        # remove element idx from Bs
        Bs = np.delete(Bs, idx)
        z_sum = np.delete(z_sum, idx)

        if plot_derivative:
            z_sum = np.gradient(z_sum, Bs)

        zs = np.append(zs, z_sum)

    num_T = len(Ts)
    num_B = len(Bs)

    X, Y = np.meshgrid(Ts, Bs)
    Z = zs.reshape(num_T, num_B).T
    if plot_derivative:
        Z = Z
    else:
        Z = np.log(Z)

    cmap = plt.get_cmap(cmap)
    data = ax.pcolormesh(X, Y, Z, cmap=cmap, shading=shading)
    fig.colorbar(data, ax=ax)

    if plot_derivative:
        data.set_clim(-50, 50)
    else:
        data.set_clim(0, np.max(Z))

    if xlim:
        ax.set_xlim(xlim)
    else:
        ax.set_xlim(0.05, np.max(Ts))

    ax.set_xlabel('Temperature (K)')
    ax.set_ylabel('Magnetic field (T)')
    if title is not None:
        ax.set_title(title)

def plot_phase_diagram(folder, ax, fig, property, title, list_of_output, quantity_column_selection=5, **args):
    dict_of_output = list_of_output[property]
    # print('dict_of_output:', dict_of_output)

    output = property
    start_column = dict_of_output['start_column']
    final_column = dict_of_output['final_column']
    quantity_size = dict_of_output['quantity size']
    n_spin = dict_of_output['n_spin']

    # which column? mx, my, mz, |m|
    quantity_column = min(quantity_column_selection, quantity_size)
    # quantity_column = quantity_size
    # print('quantity_column:', quantity_column, quantity_column_selection, quantity_size)

    ys = np.zeros(0)
    list_columns = []
    for i in range(n_spin):
        list_columns.append( start_column-1 + i*quantity_size + quantity_column )

    if quantity_column_selection == 5: list_columns = list(range(start_column, final_column))
    # print('property:', property, ', columns:', list_columns); print()
    phase_diagram_colormap(folder, color='tab:blue'  ,ax=ax, fig=fig, property = list_columns, title=title, **args)

def phase_diagram_colormap_plotall(folder, quantity_column_selection=5, **args):
    fig, axes = plt.subplots(1,4, figsize=(25,4))

    if 'toprint' in args:
        toprint = args['toprint']
    else :
        toprint = False

    if 'fig_title' in args:
        fig_title = args['fig_title']
    else :
        fig_title = None

    list_of_output = column_to_plot(folder, toprint=toprint)

    plot_phase_diagram(folder, axes[0], fig, property='mean-susceptibility',          title='Susceptibility',          list_of_output=list_of_output, quantity_column_selection=quantity_column_selection, **args)
    plot_phase_diagram(folder, axes[1], fig, property='material-mean-susceptibility', title='Susceptibility per spin', list_of_output=list_of_output, quantity_column_selection=quantity_column_selection, **args)
    plot_phase_diagram(folder, axes[2], fig, property='mean-specific-heat',           title='Specific Heat',           list_of_output=list_of_output, quantity_column_selection=quantity_column_selection, **args)
    plot_phase_diagram(folder, axes[3], fig, property='material-mean-specific-heat',  title='Specific heat per spin',  list_of_output=list_of_output, quantity_column_selection=quantity_column_selection, **args)

    # add a title to the figure
    if fig_title : fig.suptitle(fig_title, fontsize=16)
````

````python
import matplotlib.pyplot as plt

outputfile = '/content/drive/MyDrive/Colab Notebooks/vampire curieTemp/output'
# outputfile = 'vampire curieTemp/output'

temperatures = np.loadtxt(outputfile, usecols=(0))
magnetization = np.loadtxt(outputfile, usecols=(4))
susceptibility = np.loadtxt(outputfile, usecols=(8))
specific_heat = np.loadtxt(outputfile, usecols=(9))

plt.plot(temperatures, magnetization, label='magnetization')
plt.xlabel('Temperature (K)')
plt.ylabel('Magnetization')
plt.show()

plt.plot(temperatures, susceptibility, label='susceptibility')
plt.ylabel('Susceptibility')
plt.xlabel('Temperature (K)')
plt.show()

plt.plot(temperatures, specific_heat, label='susceptibility')
plt.ylabel('Specific heat')
plt.xlabel('Temperature (K)')
plt.show()
````

````python
import scipy

def mag(T, Tc, beta, C):
   if T>=Tc:
      mag = 0.0
   else:
      mag = ( 1.0 - T/Tc )**beta * C

   return mag

magv = np.vectorize(mag)

xs_fit = np.linspace(temperatures[0], temperatures[-1], 500, endpoint=True)

result = scipy.optimize.curve_fit(magv, temperatures, magnetization, p0=(200.0, 0.5, 0.1))
Tc = result[0][0]
beta = result[0][1]
C = result[0][2]
print(f"fit: Tc= {Tc:f} beta= {beta:f} \n")

base_line, = plt.plot(temperatures, magnetization, 'o')

_ = plt.plot(xs_fit, magv(xs_fit, Tc, beta, C), '-', c=base_line.get_color())
````
